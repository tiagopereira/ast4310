#### General Klein-Nishina cross section

```julia
"""
Total Klein-Nishina cross section,
normalised by Thomson cross section
"""
function sigma_kn(x)
    s = (3/4 * (1 + x) / x^3 * (2 * x * (1 + x) / (1 + 2 * x) - log(1 + 2 * x)) + 
         log(1 + 2 * x) / (2 * x) - (1 + 3 * x)/(1 + 2 * x)^2 )
    return s
end
```
