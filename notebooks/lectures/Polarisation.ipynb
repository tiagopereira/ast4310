```julia
using CairoMakie
using Printf 
```

```julia
function plot_stokes(E1, E2, ϕ1, ϕ2)
    z = range(0, stop=8π, length=200)
    Ex = E1 .* cos.(z .+ ϕ1)
    Ey = E2 .* cos.(z .+ ϕ2)
    I = E1^2 + E2^2
    Q = E1^2 - E2^2
    U = 2 * E1 * E2 * cos(ϕ1 - ϕ2)
    V = 2 * E1 * E2 * sin(ϕ1 - ϕ2)

    fig = Figure(size=(600, 400))
    ax = Axis3(
        fig[1, 1], 
        aspect=(3, 1, 1), 
        xlabel="z", 
        ylabel="Ey",
        zlabel="Ex",
        title = @sprintf("(I, Q, U, V) = (%.1f, %.1f, %.1f, %.1f)", I, Q, U, V)
    )
    lines!(ax, zeros(length(z)), Ex, Ey, color=:darkorange, linewidth=2)
    lines!(ax, z, Ex, Ey, color=:royalblue3, linewidth=2)
    lines!(ax, z, fill(maximum(Ey), length(z)), Ey,
           color=:black, linestyle=:dot, linewidth=0.8)
    lines!(ax, z, Ex, fill(minimum(Ex), length(z)),
           color=:black, linestyle=:dot, linewidth=0.8)
    xlims!(ax, 0, last(z))
    positions = collect(range(first(z), stop=last(z), step=π))
    ax.xticks = (positions, fill("", length(positions)))
    return fig
end
```

```julia
plot_stokes(1, 1, 0, 0)
```
