```julia
using CairoMakie
```

```julia
compute_slab_intensity(τ, i0, S) = i0 * exp.(-τ) .+ S * (1 .- exp.(-τ))
```

```julia
τ = logrange(0.001, 10^0.7, 100)
S = 1
fig = Figure(size=(500,300))
ax = Axis(fig[1,1])
#ax = Axis(fig[1,1], yscale=log10, xscale=log10)

for i0 in 0:0.5:2
    intensity = compute_slab_intensity(τ, i0, S)
    lines!(ax, τ, intensity)
end

lines!(ax, τ, ones(length(τ)); color=:black, linestyle=:dash)
lines!(ax, τ, τ * S; color=:black, linestyle=:dash)
ylims!(ax, 1e-3, 2)
fig
```
