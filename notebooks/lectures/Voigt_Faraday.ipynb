```julia
using CairoMakie
import SpecialFunctions: erfcx, dawson
```

```julia
voigt(a, v) = real(erfcx(a - v*im)) / sqrt(π)
faraday(a, v) = imag(erfcx(a - v*im)) / sqrt(π)
```

```julia
u = range(-10, 10, 200)
a = logrange(1e-3, 1, 4)

fig = Figure(size=(500,350))
ax = Axis(fig[1,1], xgridvisible=false, ygridvisible=false)
for ai in a
    lines!(ax, u, voigt.(ai, u), label="a=$ai")
end
axislegend(position=:rt, framevisible=false)
hidespines!(ax, :t, :r)
fig
```
