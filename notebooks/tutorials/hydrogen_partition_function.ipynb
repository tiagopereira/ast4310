```python
%matplotlib ipympl
%config InlineBackend.figure_format = 'svg'

import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u

from astropy.constants import k_B as k
```

For the hydrogen atom, we have the Rydberg formula
\begin{equation}
X_{1, s} = 13.598 \left( 1 - \frac{1}{s^2} \right) [\text{eV}]\,,
\end{equation}

and statistical weights
\begin{equation}
g_{1, s} = 2s^2\,.
\end{equation}

```python
def chi(s):
    chi_0 = 13.598*u.eV
    return chi_0 * (1 - 1/s**2)

def g(s):
    return 2*s**2

def U_i(T,s):
    return g(s) * np.exp(-chi(s)/(k * T))
```
